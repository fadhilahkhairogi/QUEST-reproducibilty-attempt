```python
import numpy as np
import matplotlib.pyplot as plt

# =========================
# DATA (arrays first)
# =========================
filters = np.array([4, 5, 6, 7, 8, 9])

# Pink bars (filter ordering time)  exponential growth
ordering_time = np.array([1, 4, 15, 60, 500, 3000])

# Blue bars (processing time)  small, sub-linear
processing_time = np.array([0.35, 0.40, 0.55, 0.70, 0.85, 1.00])

# =========================
# FIGURE SETUP
# =========================
plt.figure(figsize=(8, 6))
bar_width = 0.35

# =========================
# BAR PLOTS
# =========================
plt.bar(
    filters - bar_width/2,
    ordering_time,
    width=bar_width,
    color='#f7b6c2',
    edgecolor='black',
    hatch='x',
    label='Filter Ordering'
)

plt.bar(
    filters + bar_width/2,
    processing_time,
    width=bar_width,
    color='#9ecae1',
    edgecolor='black',
    hatch='.',
    label='Processing'
)

# =========================
# LINE PLOT (black curve)
# =========================
plt.plot(
    filters,
    ordering_time,
    color='black',
    linewidth=2.5,
    marker='o'
)

# =========================
# AXES & SCALE
# =========================
plt.yscale('log')
plt.ylim(1e-1, 5e3)   # <-- IMPORTANT: baseline at 10^-1
plt.xticks(filters, fontsize=18)
plt.yticks(fontsize=16)

plt.xlabel('Number of Filters', fontsize=22, fontweight='bold')
plt.ylabel('Time (s)', fontsize=22, fontweight='bold')
plt.title('Filter Groups\nTime Comparison', fontsize=24, fontweight='bold')

# =========================
# STYLE
# =========================
plt.grid(False)
plt.legend(fontsize=14, loc='upper left')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# =========================
# DATA (arrays first)
# =========================
filters = np.array([4, 5, 6, 7, 8, 9])


# Pink bars (filter ordering time)  exponential growth
# ordering_time = np.array([1, 4, 15, 60, 500, 3000])
x = 1.075562
# x = 0.0027
ordering_time = np.array([24*x, 120*x, 720*x, 5040*x, 40320*x, 362880*x])

print(362880*x)
# Blue bars (processing time)  small, sub-linear
processing_time = np.array([1.342988, 1.446346, 1.268216, 1.915594, 1.736519, 2.068898])

# =========================
# FIGURE SETUP
# =========================
plt.figure(figsize=(8, 6))
bar_width = 0.35

# =========================
# BAR PLOTS
# =========================
plt.bar(
    filters - bar_width/2,
    ordering_time,
    width=bar_width,
    color='#f7b6c2',
    edgecolor='black',
    hatch='x',
    label='Exhaust'
)

plt.bar(
    filters + bar_width/2,
    processing_time,
    width=bar_width,
    color='#9ecae1',
    edgecolor='black',
    hatch='.',
    label='QUEST'
)

# =========================
# LINE PLOT (black curve)
# =========================
plt.plot(
    filters,
    ordering_time,
    color='black',
    linewidth=2.5,
    marker='o'
)

# =========================
# AXES & SCALE
# =========================
plt.yscale('log')
plt.ylim(1e-1, 7e5)   # <-- IMPORTANT: baseline at 10^-1
plt.xticks(filters, fontsize=18)
plt.yticks(fontsize=16)
plt.yticks([1e1, 1e3], [r'$10^1$', r'$10^3$'])

plt.xlabel('Number of Filters', fontsize=22, fontweight='bold')
plt.ylabel('Time (s)', fontsize=22, fontweight='bold')
plt.title('Filter Groups\nTime Comparison', fontsize=24, fontweight='bold')

# =========================
# STYLE
# =========================
plt.grid(False)
plt.legend(fontsize=14, loc='upper left')

plt.tight_layout()
plt.show()
```

```python
import numpy as np

# data
x = np.array([4, 5, 6, 7, 8, 9])
y = np.array([1.342988, 1.446346, 1.268216, 1.915594, 1.736519, 2.068898])

# linear regression (y = a*x + b)
a, b = np.polyfit(x, y, 1)

# prediction
x_pred = 1
y_pred = a * x_pred + b

print("slope (a):", a)
print("intercept (b):", b)
print("prediction at x=1:", y_pred)
```

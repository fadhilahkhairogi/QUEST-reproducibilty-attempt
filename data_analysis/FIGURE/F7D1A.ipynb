```python
import matplotlib.pyplot as plt
import numpy as np

# Data (hard-coded approximations)
ratios = [0.01, 0.03, 0.05, '0.10', 0.15, '0.20', 0.25]
x = np.arange(len(ratios))

precision_vals = [1, 1, 0.5, 0.571429, 0.571429, 0.75, 0.928571428571428]
recall_vals    = [0.25, 0.5, 0.75, 1, 1, 0.75 , 0.764706]
f1_vals        = [0.571428571428571, 0.66666, 0.6, 0.72727272, 0.72727272, 0.75,0.838710]

width = 0.22

fig, ax = plt.subplots(figsize=(9,6.6))

# Bars with hatch patterns
# bars1 = ax.bar(x - width, precision_vals, width, label='Precession', hatch='///', edgecolor='black', linewidth=0.6)
# bars2 = ax.bar(x, recall_vals, width, label='Recall', hatch='\\\\\\\\', edgecolor='black', linewidth=0.6)
# bars3 = ax.bar(x + width, f1_vals, width, label='F1 Score', hatch='xx', edgecolor='black', linewidth=0.6)
bars1 = ax.bar(
    x - width,
    precision_vals,
    width,
    label='Precission',
    hatch='/',
    edgecolor='black',
    linewidth=0.6,
    color=(0.95, 0.70, 0.70, 0.8)   # merah pastel
)

bars2 = ax.bar(
    x,
    recall_vals,
    width,
    label='Recall',
    hatch='\\',
    edgecolor='black',
    linewidth=0.6,
    color=(0.70, 0.90, 0.95, 0.8)   # biru pastel
)

bars3 = ax.bar(
    x + width,
    f1_vals,
    width,
    label='F1 Score',
    hatch='x',
    edgecolor='black',
    linewidth=0.6,
    color=(0.75, 0.95, 0.75, 0.8)   # hijau pastel
)


# Lines overlay (markers)
ax.plot(x - width, precision_vals, marker='s', linestyle='-', linewidth=1.2, color='#B24F4F')
ax.plot(x, recall_vals, marker='o', linestyle='-', linewidth=1.5, color='#57a6a6')
ax.plot(x + width, f1_vals, marker='^', linestyle='-', linewidth=1.2, color='#4b8b46')

# Labels and ticks
ax.set_xlabel('Ratio', fontsize=11)
ax.set_ylabel('Score', fontsize=11,fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels([str(r) for r in ratios], fontsize=10)
ax.set_ylim(0.2, 1.02)

# Legend
# legend = ax.legend(frameon=True, fontsize=9, loc='upper left', ncol=3)
# legend.get_frame().set_edgecolor('black')
legend = ax.legend(
    frameon=True,
    fontsize=23,
    ncol=3,
    loc='upper center',
    bbox_to_anchor=(0.5, 1.22),   # dorong legend ke atas luar plotting area
)


legend.get_frame().set_edgecolor('black')
legend.get_frame().set_alpha(0.2)


# Grid and frame
ax.grid(axis='y', linestyle='--', linewidth=0.5, alpha=0.6)
ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(True)

# Subplot label/caption-like text
ax.text(0.02, -0.14, '(d) Ablation Study of Sample Ratio', transform=ax.transAxes, fontsize=11, weight='bold')

plt.tight_layout()
plt.subplots_adjust(top=0.72)
out_path = "./ablation_d.png"
plt.savefig(out_path, dpi=300, bbox_inches='tight')
plt.show()
```

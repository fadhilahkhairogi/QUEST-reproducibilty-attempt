```python
import pandas as pd


def statis(path =None, input_file = "/ordering_time_full_figure5d.txt"):
    # LOAD DATA (no header in txt)
    
    if not path:
        path = "Z0040X"
    path = '../exp_result/' + path
    input_file = path + input_file


    df = pd.read_csv(
        input_file,
        header=None,
        names=["file", "num_filters", "time"]
    )

    # ENSURE NUMERIC
    df["time"] = pd.to_numeric(df["time"], errors="coerce")
    df["num_filters"] = pd.to_numeric(df["num_filters"], errors="coerce")

    # DROP INVALID ROWS
    df = df.dropna(subset=["time", "num_filters"])

    # GROUP BY NUMBER OF FILTERS (4â€“9)
    grouped = df.groupby("num_filters")["time"]

    # COMPUTE STATISTICS
    stats = pd.DataFrame({
        "count": grouped.count(),
        "sum": grouped.sum(),
        "mean": grouped.mean(),
        "median": grouped.median(),
        "q1": grouped.quantile(0.25),
        "q3": grouped.quantile(0.75),  # same as median, kept for paper clarity
        "stdev": grouped.std(),
        "variance": grouped.var(),
        "maximum": grouped.max(),
        "minimum": grouped.min()
    })

    # OUTPUT
    # output_file = "ordering_time_full_figure5d_stats.csv"
    # stats.to_csv(output_file)

    print("=== Figure 5(d) Ordering Time Statistics ===")
    print(
        stats
        .round(6)
        .to_string()
    )
    # print(stats)
    # print(f"\nSaved to: {output_file}")
```

```python
statis()
```
